```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

```python
calender = pd.read_csv("calendar.csv")
sell_price = pd.read_csv("sell_prices.csv")
sales_train_val = pd.read_csv("sales_train_validation.csv")
smp_submission = pd.read_csv("sample_submission.csv")
```

```python
#       &  
plt.figure(figsize = (90,10))
dat = sales_train_val.iloc[:,6:].sum(axis=0)
event_date = tuple(zip(calender['d'][~calender['event_name_1'].isnull()],calender['event_name_1'][~calender['event_name_1'].isnull()]))
holiday = ['PresidentDay', 'MemorialDay', 'IndependenceDay', 'LaborDay', 'ColumnbusDay', 'VeteransDay', 'Thanksgiving', 'Christmas', 'NewYear', 'MartinLutherKingDay']

plt.plot(dat)
for i, (d, event_name) in enumerate(event_date):
    if d not in dat.index:
        break
    plt.axvline(x=d, color='blue')
    if event_name in holiday:
        plt.axvline(x=d, color='red')
    plt.text(x=d, y=[55000, 54000, 53000, 52000, 51000][i%5], s=event_name)
plt.ylim((15000, 60000))
plt.show()
```
